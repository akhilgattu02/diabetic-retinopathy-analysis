```python
import cv2

def read_image(path):
    img = cv2.imread(path)
    return img

def show_image(window_name, img):
    cv2.imshow(window_name, img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

img = read_image('A. Segmentation/1. Original Images/a. Training Set/IDRiD_01.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_threshold = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY_INV)
show_image('Thresholded Image', img_threshold[1])
```
